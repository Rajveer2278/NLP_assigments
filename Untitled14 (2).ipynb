```python
import pandas as pd

data1 = pd.read_pickle('corpus.pkl')
data1
```

```python
# Create quick lambda functions to find the polarity and subjectivity of each routine

from textblob import TextBlob
#TextBlob is a Python library for processing textual data.
#Polarity:- how + or - a sentence is.
#Subjectivity:- how (subjective) or (objective) a sentence is.

pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity
#lambda arguments: expression
#how factual or opinion-based the text is

data1['polarity'] = data1['transcript'].apply(pol)
data1['subjectivity'] = data1['transcript'].apply(sub)
data1

#-----------> Polarity (ranges from -1 to +1):
# -1.0: Very negative
# 0.0: Neutral
# +1.0: Very positive
#-----------> Subjectivity (ranges from 0 to 1):
# 0.0: Completely objective (fact-based)
# 1.0: Completely subjective (opinion-based)
```

```python
top_positive = data1.sort_values('polarity', ascending=False)
print(top_positive[['full_name', 'polarity']].head(5))
```

```python
most_subjective = data1.sort_values('subjectivity', ascending=False)
print(most_subjective[['full_name', 'subjectivity']].head(5))
```

The majority of comedians fall within the 0.0 to 0.2 range, meaning their content is neutral to mildly positive.

Range of 0 to 0.1 -They may use sarcasm, dark humor, or social commentary which reflects in their lower polarity.like Dave, Bill,Joe etc

Some comedians have notably different styles  Michelle Buteau, and Nate Bargatze have polarity scores above the mean, implying their content is consistently positive.

```python
import numpy as np
import math

def split_text(text, n=10):
    '''Splits text into n nearly equal parts while keeping words intact.'''

    words = text.split()  # Tokenize into words (Splits the text into a list of words)
    total_words = len(words) #how many words are in the text
    size = math.ceil(total_words / n) # ensure no word loss happens even if some chunks are of different sizes (how many words should go into each chunk)

    # Create chunks by grouping words
    split_list = [' '.join(words[i:i+size]) for i in range(0, total_words, size)]

    return split_list
```

```python
# Let's create a list to hold all of the pieces of text
list_piecess = []
for t in data1.transcript:
    split = split_text(t)
    list_piecess.append(split)

list_piecess
```

1.Modify the number of sections the comedy routine is split into and see how the charts over time change.

```python
import numpy as np, pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from textblob import TextBlob

#takes piece of text and return its polarity
def get_polarity(text): return TextBlob(text).sentiment.polarity

def analyze_sentiment_progression(n_values):
    comedians = data1['full_name'].unique()#unique comedian
    fig = make_subplots(
        rows=(rows := 7), cols=3, subplot_titles=comedians)

#Loops through each comedian by index and name.
    for i, comedian in enumerate(comedians):
        row, col = divmod(i, 3); row += 1; col += 1  #determine the row and column position for subplot

        polarities = {n: [] for n in n_values} #dictionary
        for n in n_values:
            for t in data1[data1['full_name'] == comedian]['transcript']:
                pieces = split_text(t, n)          #Splits transcript into n nearly equal parts
                polarities[n].append([get_polarity(p) for p in pieces]) #stores the list of polarities

        for n, p_list in polarities.items():
            if p_list:
                avg = np.mean(p_list, axis=0)
                fig.add_trace(go.Scatter(
                    x=list(range(len(avg))), y=avg,
                    mode='lines+markers', name=f"{comedian} (n={n})",
                    legendgroup=comedian), row=row, col=col)



    fig.update_layout(height=300*rows, width=1200,
                      title_text="Sentiment Progression of Comedians", showlegend=True)
    fig.show()
```

n=5 -----> Gives a macro view of tone changes across the show; smooth overall trend(Fewer, broader chunks)

n=10 -----> Balanced view of sentiment changes; highlights notable shifts(Moderate detail)

n=20 or 30 ------> Reveals micro-shifts in tone; may show mood swings, sarcasm, punchline effects(Finer segmentation)
